# Objective  
- Removing the following columns: [“date”, “lights”]. 
- Normalize the dataset using the MinMaxScaler
- The target variable is “Appliances”. 
- Use a 70-30 train-test set split with a  random state of 42 (for reproducibility). 
- Run a multiple linear regression using the training set and evaluate your model on the test set.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Models
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

# Metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import  mean_squared_error
```

```python
# Drop Date and Light
new_energy_df = energy_df.drop(columns=['date','lights'])
new_energy_df.info()
```

```python
# Normalise the dataset to a common scale using the min max scaler
scaler = MinMaxScaler()
normalised_energy_df = pd.DataFrame(scaler.fit_transform(new_energy_df), columns=new_energy_df.columns)
```

```python
# Split dataset into feature and target
features_df = normalised_energy_df.drop(columns=['Appliances'])
appliances_target = normalised_energy_df['Appliances']
```

```python
# Split our dataset into the training and testing dataset with a ratio of 70:30 respectively
x_train, x_test, y_train, y_test = train_test_split(features_df, appliances_target, test_size=0.3, random_state=42)
```

### Train the Machine learning model

```python
# Run Linear model
linear_model = LinearRegression()

# train the model 
linear_model.fit(x_train, y_train)

# Obtain predictions
predicted_values = linear_model.predict(x_test)
```

### Visualizing the predicted value

```python
df_pred = pd.DataFrame({
    "true_value":y_test,
    "predicted_value":predicted_values
})
```

```python
df_pred['x_axis'] = range(df_pred.shape[0])
```

## Metrics

```python
# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, predicted_values)
print('Mean Absolute Error: {}'.format(round(mae, 3)))
```

```python
# R-Squared
r2 = r2_score(y_test, predicted_values)
print('R-Squared: {}'.format(round(r2, 3)))
```

```python
# Residual Sum of Squares (RSS)
rss = np.sum(np.square(y_test - predicted_values))
print('Residual Sum of Squares: {}'.format(round(rss, 3)))
```

```python
# Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('Root Mean Square Error: {}'.format(round(rmse, 3)))
```

## Regression

```python
# Lasso regression
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
# Obtain predictions
predicted_lasso_values = lasso_reg.predict(x_test)
# Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predicted_lasso_values))
print('Root Mean Square Error: {}'.format(round(rmse, 3)))
```

```python
# Ridge regression
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
# Obtain predictions
predicted_ridge_values = ridge_reg.predict(x_test)
# Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predicted_ridge_values))
print('Root Mean Square Error: {}'.format(round(rmse, 3)))
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
